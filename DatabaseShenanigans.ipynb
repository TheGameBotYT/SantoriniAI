## Select one filtered

```python
conn = sqlite3.connect('SantoriniAI.db')
select_query = '''
SELECT
    s.Action1,
    s.Action2,
    s.Action3,
    s.Action4,
    s.Action5,
    s.Action6,
    s.Action7,
    s.Action8
FROM Simplorini as s
WHERE (s.phase = ?)
AND (s.CurrentPosition = ?)
AND (s.OpponentPosition = ?)
AND (s.Build0 = ?)
AND (s.Build1 = ?)
'''
cur = conn.cursor()
cur.execute(select_query, (0, 4, 1, 0, 0))
data = cur.fetchall()
conn.commit()
```

```python
create_table_query = '''
CREATE TABLE
    Simplorini(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    phase INTEGER NOT NULL,
    CurrentPosition INTEGER NOT NULL,
    OpponentPosition INTEGER NOT NULL,
    Build0 INTEGER NOT NULL,
    Build1 INTEGER NOT NULL,
    Build2 INTEGER NOT NULL,
    Build3 INTEGER NOT NULL,
    Build4 INTEGER NOT NULL,
    Build5 INTEGER NOT NULL,
    Build6 INTEGER NOT NULL,
    Build7 INTEGER NOT NULL,
    Build8 INTEGER NOT NULL,
    Action1 REAL,
    Action2 REAL,
    Action3 REAL,
    Action4 REAL,
    Action5 REAL,
    Action6 REAL,
    Action7 REAL,
    Action8 REAL)
'''
conn = sqlite3.connect('SantoriniAI.db')
cur = conn.cursor()
cur.execute(create_table_query)
conn.commit()
conn.close()
```

```python
query = """
INSERT INTO Simplorini (id, phase, CurrentPosition, OpponentPosition,
                Build0, Build1, Build2, Build3, Build4, Build5, Build6, Build7, Build8,
                Action1, Action2, Action3, Action4, Action5, Action6, Action7, Action8)
                VALUES(Null, ?, ?, ?,
                ?, ?, ?, ?, ?, ?, ?, ?, ?,
                ?, ?, ?, ?, ?, ?, ?, ?)
"""
conn = sqlite3.connect('SantoriniAI.db')
cur = conn.cursor()
cur.executemany(query, tuple(data_list))
conn.commit()
conn.close()
```
